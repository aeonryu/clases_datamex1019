# Datos

```python
import warnings
warnings.simplefilter('ignore')

import pandas as pd
from sklearn.datasets import load_diabetes, load_wine

data=load_diabetes()
diabetes=pd.DataFrame(data.data, columns=data.feature_names)
diabetes['target']=data.target
diabetes.head()
```

```python
data=load_wine()
vino=pd.DataFrame(data.data, columns=data.feature_names)
vino['target']=data.target
vino.head()
```

```python
from sklearn.model_selection import train_test_split

X_train_d, X_test_d, y_train_d, y_test_d=train_test_split(diabetes.drop('target', axis=1), diabetes.target)

X_train_v, X_test_v, y_train_v, y_test_v=train_test_split(vino.drop('target', axis=1), vino.target)
```

```python
# SGDR

from sklearn.linear_model import SGDRegressor as SGDR
sgdr=SGDR()
sgdr.fit(X_train_d, y_train_d)
```

```python
train_score=sgdr.score(X_train_d, y_train_d)
test_score=sgdr.score(X_test_d, y_test_d)
print (train_score, test_score)
```

```python
# SGDC

from sklearn.linear_model import SGDClassifier as SGDC
sgdc=SGDC()
sgdc.fit(X_train_v, y_train_v)
```

```python
train_score=sgdc.score(X_train_v, y_train_v)
test_score=sgdc.score(X_test_v, y_test_v)
print (train_score, test_score)
```

# H2O (ML for dummies)

```python
import h2o
from h2o.automl import H2OAutoML

h2o.init()
```

```python
url='https://s3.amazonaws.com/erin-data/higgs/'

train=h2o.import_file(url+'higgs_train_10k.csv')
test=h2o.import_file(url+'higgs_test_5k.csv')
```

```python
X=train.columns
y='response'
X.remove(y)
```

```python
train[y]=train[y].asfactor()
test[y]=test[y].asfactor()
```

```python
# modelo

aml=H2OAutoML(max_models=20, seed=1)
aml.train(x=X, y=y, training_frame=train)
```

```python
lb=aml.leaderboard
lb.head(rows=lb.nrows)
```

```python
aml.leader
```

```python
preds=aml.leader.predict(test)
preds
```

# Stacking

```python
df=pd.read_csv('data/diamonds.csv')
df.head()
```

```python
X=df.drop(columns=['Unnamed: 0', 'price', 'cut', 'color', 'clarity'])
y=df.price
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y)
```

```python
n_df=pd.DataFrame()
```

```python
n_df['xgbr']=xgbr.fit(X_train, y_train).predict(X_test)
```

```python
n_df['lbgmr']=lgbmr.fit(X_train, y_train).predict(X_test)
```

```python
n_df['ctr']=ctr.fit(X_train, y_train).predict(X_test)
```

```python
n_df['gbr']=gbr.fit(X_train, y_train).predict(X_test)
```

```python
n_df.head()
```

```python
from sklearn.linear_model import LinearRegression as LR

linreg=LR()
linreg.fit(n_df.values, y_test)
```

```python
linreg.score(n_df.values, y_test)
```

```python
xgbr.score(X_test, y_test)
```

```python
lgbmr.score(X_test, y_test)
```

```python
ctr.score(X_test, y_test)
```

```python
gbr.score(X_test, y_test)
```

```python
# mlxtend

from mlxtend.regressor import StackingRegressor

metamodelo=StackingRegressor(regressors=[xgbr, gbr, lgbmr, ctr], meta_regressor=linreg)
metamodelo.fit(X_train, y_train)
metamodelo.score(X_test, y_test)
```
