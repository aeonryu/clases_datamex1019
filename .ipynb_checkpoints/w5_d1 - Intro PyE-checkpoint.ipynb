```python
# probs (dado)
def par(x):
    return (x%2==0)
```

```python
sample_space=[1,2,3,4,5,6]
caras=len(sample_space)
pares=len([e for e in sample_space if par(e)])
```

```python
pares_prob=pares/caras
pares_prob
```

```python
# dist discretas
# binomial

from scipy.stats import binom
#binom.pmf(k, n, p)
binom.pmf(3, 10, 1/6) # tres veces el 3 si tiro 10 veces?
```

```python
binom.cdf(3, 10, 1/6)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
x=[i for i in range(100)]
y=[binom.pmf(e, 100, 1/6) for e in x]
plt.plot(x, y);
```

```python
x=[i for i in range(10)]
y=[binom.cdf(e, 10, 1/6) for e in x]
plt.plot(x, y);
```

```python
n, p = 10, 1/6
media, var, sime, curt = binom.stats(n, p, moments='mvsk')
print (media, var, sime, curt)
```
