```python
import asyncio
```

```python
urls = [
    'https://s3-eu-west-1.amazonaws.com/ih-materials/uploads/data-static/documents/breakfast.jpg',
    'https://s3-eu-west-1.amazonaws.com/ih-materials/uploads/data-static/documents/forbidden',
    'https://s3-eu-west-1.amazonaws.com/ih-materials/uploads/data-static/documents/the-html5-breakfast-site.html'
]
```

```python
async def main():
    loop=asyncio.get_event_loop()
    futuro=[loop.run_in_executor(None, requests.get, url) for url in urls]
    
    for res in await asyncio.gather(*futuro):
        print (res.status_code)
        
loop=asyncio.get_event_loop()
loop.run_until_complete(main())
```

```python
# bs4+selenium

from selenium import webdriver
from bs4 import BeautifulSoup as bs
```

```python
def get_page(url):
    driver=webdriver.Firefox()
    driver.get(url)
    html=driver.page_source
    try:
        soup=bs(html, 'html.parser')
    except:
        print ('Error en la sopa.')
    driver.quit()
    return soup
```

```python
soup=get_page('https://www.ironhack.com/es')
soup
```

```python
# con clase

class PageGetter:
    driver=False
    def __init__(self, browser='firefox'):
        if browser=='firefox':
            self.driver=webdriver.Firefox()
        elif browser=='chrome':
            self.driver=webdriver.Chrome()
        else:
            raise Exception ('Cambia el buscador')
    
    def get_page(self, url):
        if not self.driver:
            raise Exception ('Escoge buscador...')
        self.driver.get(url)
        html=self.driver.page_source
        soup=bs(html, 'html.parser')
        return soup
    
    def close(self):
        self.driver.quit()
        
```

```python
pg=PageGetter('firefox')
soup=pg.get_page('https://www.ironhack.com/es')
soup
```

```python
url='https://threatbutt.com/map/'
```

```python
driver=webdriver.Firefox()

while 1:
    driver.get(url)
    time.sleep(3)
    cuenta=0
    while cuenta<1000:
        elemento=driver.find_element_by_id('attackdiv')
        t=elemento.text.split('\n')[-1]
        pred=t.split('(')[0].strip()
        ip_pred=t.split('(')[1].split(')')[0]
        prey=t.split('(')[1][-4:].strip()
        ip_prey=t.split('(')[2].split(')')[0]
        
        print ({'predator':pred, 'ip_predator':ip_pred,
                'prey':prey, 'ip_prey':ip_prey})
        
        cuenta+=1
    
```

```python
import xmltodict
import re

def getxml():
    url = 'http://www.ssn.unam.mx/rss/ultimos-sismos.xml'
    data=requests.get(url).content
    return xmltodict.parse(data)
```

```python
d=getxml()
d['rss']['channel']['item'][0]['description'].split('>')[1:4]
```

```python
def earth():
    d=getxml()
    
    for i in range(len(d['rss']['channel']['item'])):

        data_get=d['rss']['channel']['item'][i]['description'].split('>')[1:4]
        #print (data_get)
        fecha=re.findall('\d+-\d+-\d+', data_get[0])
        hora=re.findall('\d+:\d+:\d+', data_get[0])
        lat=re.findall('-?\d+.\d+', data_get[1])
        lon=re.findall('-?\d+.\d+', data_get[1])
        prof=re.findall('\d+.\d+ \w+', data_get[2])
        inten=float(d['rss']['channel']['item'][i]['title'].split(',')[0])
        
        data={'Fecha':fecha[0], 'Hora':hora[0],
              'Latitud':float(lat[0]), 'Longitud':float(lon[1]),
              'Profundidad':prof[0], 'Intensidad':inten}
        print (data)
        
```

```python
earth()
```
